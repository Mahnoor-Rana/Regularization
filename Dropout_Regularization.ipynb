# Dropout Regularization In Deep Neural Network


```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sonar.all-data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['R'].value_counts()
```

```python
X = df.drop('R',axis = 1)
y = df['R']
y.head()
```

```python
y = pd.get_dummies(y,drop_first=True)
y.sample(5)
```

```python
y.value_counts()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=1)
```

# Using Deep learning model

## Model without Dropout Layer

```python
import tensorflow as tf 
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(200,input_dim=60,activation = 'relu'),
    keras.layers.Dense(20,activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')

])
model.compile(optimizer ='adam',
             loss = 'binary_crossentropy',
             metrics= ['accuracy'])
model.fit(X_train,y_train,epochs= 100, batch_size = 5)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report ,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
model_1 = keras.Sequential([
    keras.layers.Dense(100, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model_1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_1.fit(X_train, y_train, epochs=100, batch_size=5)
```

```python
model_1.evaluate(X_test,y_test)
```

```python
y_pred = model_1.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test,y_pred))
```
